1. Prove that $A \subseteq A$

Let 𝐴 be a set and let 𝑎 ∈ 𝐴. Then 𝑎 ∈ 𝐴. So, 𝑎 ∈ 𝐴 → 𝑎 ∈ 𝐴 is true. Since 𝑎 was
an arbitrary element of 𝐴, ∀𝑥(𝑥 ∈ 𝐴 → 𝑥 ∈ 𝐴) is true. Therefore, 𝐴 ⊆ 𝐴

2. Prove that if $B \subset A$ then $A \cup B = A$

Let B⊂A. Since A⊂A, we have B∪A⊂A. Clearly, B⊂A∪B. Hence A∪B=A

3. Prove that $A \cup B = B \cup A$
